```python
pattern = r'\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s(?:am|pm)\s-\s'
```

```python
dates = [d.replace('\u202f', ' ') for d in dates] # remove \u202f...
```

```python
import re

users = []
messages = []

for i in df['user_message']:
    entry = re.split(r'([^:]+):\s', i, maxsplit=1)   # split only at first colon
    
    if len(entry) > 2:   # user + message exists
        users.append(entry[1].strip())
        messages.append(entry[2].strip())
    else:  # system notification
        users.append("group notification")
        messages.append(entry[0].strip())

df['users'] = users
df['messages'] = messages
```

```python
period = []
for hour in df[['day_name','hour']]['hour']:
    if hour == 23:
        period.append(str(hour) + '-' + str('00'))
    elif hour ==0:
        period.append(str('00') + '-' + str(hour+1))
    else:
        period.append(str(hour) + '-' + str(hour+1))
    
```

```python
df['period'] = period
```
